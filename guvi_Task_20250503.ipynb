```python
import math 

def Calc(**maths):
  if maths['operation']=='+':
    return float(maths['a'])+float(maths['b'])
  elif maths['operation']=='-':
    return float(maths['a'])-float(maths['b'])
  elif maths['operation']=='sqr':
    return int(maths['a'])**2
  elif maths['operation']=='/':
    return float(maths['a'])/float(maths['b'])
  elif maths['operation']=='sqrt':
    return float(maths['a'])**0.5
  elif maths['operation']=='log':
    return int(math.log(int(maths['a'],10)))
  elif maths['operation']=='*':
    return float(maths['a'])*float(maths['b'])
  elif maths['operation']=='**':
    return float(maths['a'])**float(maths['b'])
def new_maths(file_name):
  math_operations=[]
  with open(file_name,'r') as maths_file:
    for i in maths_file.readlines():
      math_operations.append(i)
  math_operations=[i.strip() for i in math_operations]
  math_operations=[i.split() for i in math_operations]

  for i,val in enumerate(math_operations):

      if len(val)==3:
        res=Calc(a=val[0],operation=val[1],b=val[2])
        val.append('= '+str(res))
      elif len(val)==2:
        res=Calc(a=val[0],operation=val[1])
        val.append('= '+str(res))

  return math_operations
 
math_operations=[]
with open('D:/PROFESSIONAL/GUVI/GUVI_TRAINING/Python_Coding/guvi_file_input.txt','r') as maths_file:
  for i in maths_file.readlines():
    print(i)

new_maths('D:/PROFESSIONAL/GUVI/GUVI_TRAINING/Python_Coding/guvi_file_input.txt')
Results=new_maths('D:/PROFESSIONAL/GUVI/GUVI_TRAINING/Python_Coding/guvi_file_input.txt')
Results
with open('D:/PROFESSIONAL/GUVI/GUVI_TRAINING/Python_Coding/guvi_file_input.txt','w') as maths_file:
    for i in Results:
      s=" ".join(i)
      maths_file.write(str(s)+'\n')

```
